```python
#Latihan 1
class Mahasiswa:
    def __init__(self, nama, nim, jurusan, angkatan):
        self.nama = nama
        self.nim = nim
        self.jurusan = jurusan
        self.angkatan = angkatan

    def tampilkan_biodata(self):
        print("\n=== Biodata Mahasiswa ===")
        print(f"Nama      : {self.nama}")
        print(f"NIM       : {self.nim}")
        print(f"Jurusan   : {self.jurusan}")
        print(f"Angkatan  : {self.angkatan}")

# Input dari user
print("Masukkan data mahasiswa:")
nama = input("Nama: ")
nim = input("NIM: ")
jurusan = input("Jurusan: ")
angkatan = input("Angkatan: ")

mahasiswa = Mahasiswa(nama, nim, jurusan, angkatan)

mahasiswa.tampilkan_biodata()
```

```python
class Mahasiswa:
    def __init__(self):
        self.__nama = None
        self.__nim = None
        self.__jurusan = None

    def get_nama(self):
        return self.__nama

    def set_nama(self, nama):
        if nama:
            self.__nama = nama
        else:
            print("Nama tidak boleh kosong!")

    def get_nim(self):
        return self.__nim

    def set_nim(self, nim):
        if nim.isdigit():
            self.__nim = nim
        else:
            print("NIM harus berupa angka!")

    def get_jurusan(self):
        return self.__jurusan

    def set_jurusan(self, jurusan):
        if jurusan:
            self.__jurusan = jurusan
        else:
            print("Jurusan tidak boleh kosong!")
    def input_biodata(self):
        print("Masukkan biodata mahasiswa:")
        nama = input("Nama: ")
        self.set_nama(nama)
        nim = input("NIM: ")
        self.set_nim(nim)
        jurusan = input("Jurusan: ")
        self.set_jurusan(jurusan)
        print("\nBiodata berhasil disimpan!")
    def tampilkan_biodata(self):
        if self.get_nama() and self.get_nim() and self.get_jurusan():
            print("\nBiodata Mahasiswa:")
            print(f"Nama    : {self.get_nama()}")
            print(f"NIM     : {self.get_nim()}")
            print(f"Jurusan : {self.get_jurusan()}")
        else:
            print("\nBiodata belum lengkap! Harap lengkapi semua data.")

# Program utama
def main():
    mahasiswa = Mahasiswa()
    while True:
        print("\nMenu:")
        print("1. Input Biodata")
        print("2. Tampilkan Biodata")
        print("3. Keluar")
        pilihan = input("Pilih opsi (1/2/3): ")

        if pilihan == "1":
            mahasiswa.input_biodata()
        elif pilihan == "2":
            mahasiswa.tampilkan_biodata()
        elif pilihan == "3":
            print("Program selesai. Terima kasih!")
            break
        else:
            print("Pilihan tidak valid! Harap pilih 1, 2, atau 3.")

# Menjalankan program utama
if __name__ == "__main__":
    main()
```
